```python
#### Data Processing ####
import pandas as pd

import matplotlib.pyplot as plt
#import matplotlib.animation as animation

from sklearn.decomposition import PCA
from sklearn import preprocessing

#### Plot ####
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import Scatter, Figure, Layout, Marker, Scatter3d
import colorlover as cl
init_notebook_mode(connected=True)
```

```python
eco2mix_2017 = pd.read_csv("eco2mix_data_2017.csv",index_col='Date&Time')
eco2mix_2017= eco2mix_2017.drop('Unnamed: 0', 1)
eco2mix_2017.index = pd.to_datetime(eco2mix_2017.index)
eco2mix_2017['Date'] = eco2mix_2017.index.date

eco2mix_2017daily = eco2mix_2017.groupby(eco2mix_2017['Date']).mean()

cols = ['Fioul-Global','Charbon-Global','Gaz-Global','Hydraulique-Global','Nucléaire-Global','Solaire-Global','Eolien-Global','Autres-Global','Pompage-Global','Solde-Global']

eco2mix_2017daily = eco2mix_2017daily[cols]

x = eco2mix_2017daily.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
eco2mix_2017daily = pd.DataFrame(x_scaled,columns=cols,index=eco2mix_2017daily.index)

eco2mix_2017daily['Month'] = [item.month for item in eco2mix_2017daily.index]
eco2mix_2017daily.head()
```

```python
pca = PCA(n_components=10)
pca.fit(eco2mix_2017daily.iloc[:,:-1])
print(pca.explained_variance_ratio_)  

plt.plot(range(len(pca.explained_variance_ratio_)),pca.explained_variance_ratio_)
plt.show()
```

## 2D PCA

```python
pca = PCA(n_components=2)
pca.fit(eco2mix_2017daily.iloc[:,:-1])

Xpca = pca.fit_transform(eco2mix_2017daily.iloc[:,:-1])
```

```python
data_pca = pd.DataFrame(Xpca[:,0:2], columns=['X0','X1'])
data_pca['Date'] = eco2mix_2017daily.index

plt.plot(data_pca['X0'] , data_pca['X1'])
plt.show()
```

```python
data_pca = pd.DataFrame(Xpca[:,0:2], columns=['X0','X1'])
data_pca['Month'] = list(eco2mix_2017daily['Month'])
data_pca['Date'] = list(eco2mix_2017daily.index)


seas = ['Winter' for item in range(len(data_pca.index[:59]))] + ['Spring' for item in range(len(data_pca.index[59:151]))] + ['Summer'for item in range(len(data_pca.index[151:243]))] + ['Autumn' for item in range(len(data_pca.index[243:334]))] + ['Winter' for item in range(len(data_pca.index[334:]))]
data_pca['Season'] = seas
```

```python
for idx in range(1,13):
    plt.scatter(data_pca.loc[data_pca['Month']==idx]['X0'] , data_pca.loc[data_pca['Month']==idx]['X1'] ,label=str(idx)) 
plt.legend()
plt.show()
```

```python
for season in ['Winter' , 'Spring' , 'Summer' , 'Autumn']:
    plt.scatter(data_pca.loc[data_pca['Season']==season]['X0'] , data_pca.loc[data_pca['Season']==season]['X1'] ,label=season) 
plt.legend()
plt.show()
```

```python
data_pca.to_csv('PCA2017.csv')
```

### 3D PCA

```python
pca = PCA(n_components=3)
pca.fit(eco2mix_2017daily.iloc[:,:-1])

Xpca = pca.fit_transform(eco2mix_2017daily.iloc[:,:-1])
```

```python
data_pca = pd.DataFrame(Xpca[:,0:3], columns=['X0','X1','X2'])
data_pca['Month'] = list(eco2mix_2017daily['Month'])
data_pca['Date'] = list(eco2mix_2017daily.index)


seas = ['Winter' for item in range(len(data_pca.index[:59]))] + ['Spring' for item in range(len(data_pca.index[59:151]))] + ['Summer'for item in range(len(data_pca.index[151:243]))] + ['Autumn' for item in range(len(data_pca.index[243:334]))] + ['Winter' for item in range(len(data_pca.index[334:]))]
data_pca['Season'] = seas
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for season in ['Winter' , 'Spring' , 'Summer' , 'Autumn']:
    ax.scatter(data_pca.loc[data_pca['Season']==season]['X0'] , data_pca.loc[data_pca['Season']==season]['X1'], data_pca.loc[data_pca['Season']==season]['X2'] ,label=season) 
plt.legend()
plt.show()
```

```python
#### Plot

iplot({
    'data': [Scatter3d(x = data_pca.loc[data_pca['Season']==season]['X0'],
                       y = data_pca.loc[data_pca['Season']==season]['X1'],
                       z = data_pca.loc[data_pca['Season']==season]['X2'],
                       name=season,
                       mode='markers')
                       #marker=dict(size=12,line=dict(color='rgba(217, 217, 217, 0.14)',width=0.5)
             for season in ['Winter' , 'Spring' , 'Summer' , 'Autumn'] 
    ],
    'layout': Layout()
}, show_link=False)
```
